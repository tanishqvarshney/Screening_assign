# Screening Assignment

```python
search_txt = "placement"
  
replace_txt = "screening"
  

with open(r'example.txt', 'r') as file:
  
    data = file.read()
    data = data.replace(search_txt, replace_txt)

with open(r'example.txt', 'w') as file:
    file.write(data)  
print("placement got replaced by screening")
```

# Use of abstract class, multiple inheritance and decorator

```python
from abc import ABC, abstractmethod
class Footballer(ABC):
 
    def move(self):
        pass
class Cristiano_ronaldo(Footballer):
 
    def move(self):
        print("5 time Ballon d'or winner")
class Leo_messi(Footballer):
 
    def move(self):
        print("7 time Ballon d'or winner")
class Johan_Cruyff(Footballer):
 
    def move(self):
        print("3 time Ballon d'or winner")
         
# Driver code
F = Cristiano_ronaldo()
F.move()
 
F = Leo_messi()
F.move()
 
F = Johan_Cruyff()
F.move()
```

```python
class hulk:
    def smash(self):
        return "My name is Thor"
class banner:
    def speak(self):
        return "I am the Strongest Avenger"
class smarthulk(hulk, banner):
    pass
s1 = smarthulk()
print(s1.smash(), "and", s1.speak())
```

```python
import time
import math
 
def calculate_time(func):
    def inner1(*args, **kwargs):
        begin = time.time()
        func(*args, **kwargs)
        end = time.time()
        print("Total time taken in : ", func.__name__, end - begin)
 
    return inner1
 
@calculate_time
def factorial(num):
    time.sleep(2)
    print(math.factorial(num))

factorial(5)
```
